```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import re, TfidfVectorizer
```

```python
df  = pd.read_csv("./archive/id_reviews.csv")
df.head()'
```

```python
def create_sentiment(rating):
    if rating == 1 or rating ==2:
        return -1
    elif rating == 4 or rating ==5:
        return 1
    else: return 0

def clean_data(review):
    no_punc = re.sub(r'[^\w\s]', '', review)
    no_digits = ''.join([i for i in no_punc if not i.isdigit()])

    return no_digits
    

df['Sentiment'] = df['Score'].apply(create_sentiment)
df['Reviews'] = df['Reviews'].apply(clean_data)
```

```python
df.head()
```

```python
tf = TfidfVectorizer(strip_accents = None,
                        lowercase=False,
                        preprocessor=None)
tfidf_matrix = tf.fit_transform(df['Reviews'])
feature_names = tf.get_feature_names_out()


for doc in range(len(df['Reviews'])):
    print("review: ", doc, "sentiment: ", df['Sentiment'][doc])
    feature_index = tfidf_matrix[doc,:].nonzero()[1]
    tfidf_scores = zip(feature_index, [tfidf_matrix[doc, x] for x in feature_index])

    for w, s in [(feature_names[i], s) for (i, s) in tfidf_scores]:
        print( w, s)
    print()
```
